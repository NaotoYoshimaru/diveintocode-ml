**【問題1】富士山を超える最小の折る回数**<br>
厚さが「富士山(3776m)」を超えるのは何回紙を折った時か計算するプログラムを作成してください。

```
fuji_height = 3776
paper_thickness = 0.00008
thickness_list = [paper_thickness]

while fuji_height > paper_thickness:
  paper_thickness = paper_thickness*2
  thickness_list.append(paper_thickness)
  print(thickness_list)
number = len(thickness_list) -1
print("{}回紙を折ると富士山を超える".format(number))
```
