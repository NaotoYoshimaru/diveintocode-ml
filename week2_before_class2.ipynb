##**【問題1】乱数の作成**
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
\begin{bmatrix}
1.0 & 0.8 \\
0.8 & 1.0
\end{bmatrix}
$$
<br><br>
多次元正規分布(μ,Σ)に従う正規乱数を生成する関数として


```
numpy.random.multivariate_normal(mean, cov, size, check_valid, tol)
```
が用意されている。引数には<br>
* loc: 生成したい正規分布の平均値ベクトルを指定
* scale: 生成したい正規分布の分散共分散行列を指定
* size: 生成する乱数の数を指定(default=1)
* check_valid: 分散共分散行列の半正定値チェック有無(warn/raise/ignoreを指定。オプション)
* tol: 半正定値チェックの許容誤差(オプション)




````
import numpy as np

np.random.seed(0)
mu = [-3, 0]
sigma = [[1.0, 0.8],[0.8, 1.0]]

values = np.random.multivariate_normal(mu, sigma, 500)
values
````

##**【問題2】散布図による可視化**
問題1で作成したデータ点を散布図により可視化してください。


散布図はmatplotlibのplt.scatter()を使うことで描けます。

````
import matplotlib.pyplot as plt
plt.scatter(values[:,0], values[:,1])
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
````

##**【問題3】ヒストグラムによる可視化**
問題1で作成したデータをヒストグラムにより可視化してください。


ヒストグラムはplt.hist()を使うことで描けます。


matplotlib.pyplot.hist — Matplotlib 3.1.1 documentation


各次元ごとに作成し、2つのヒストグラムとしてください。


それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。

````
plt.hist(values[:,0], bins=50)
plt.title("histogram of x1")
plt.xlabel("x1")
plt.ylabel("trequency")
plt.xlim(-6, 3)
plt.show()
````

````
plt.hist(values[:,1],bins=50)
plt.title("histogram of x2")
plt.xlabel("x2")
plt.ylabel("trequency")
plt.xlim(-6, 3)
plt.show()
````

##**【問題4】データの追加**
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
$$
\begin{bmatrix}
1.0 & 0.8 \\
0.8 & 1.0
\end{bmatrix}
$$

````
np.random.seed(0)
mu = [0, -3]
sigma = [[1.0, 0.8],[0.8, 1.0]]

values_new = np.random.multivariate_normal(mu, sigma, 500)
values_new
````

````
plt.scatter(values[:,0], values[:,1], label="0")
plt.scatter(values_new[:,0], values_new[:,1], label="1")
plt.title("scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
````

##**【問題5】データの結合**
結合はnp.concatenate()やnp.vstack()を使うことで行えます




````
merged_data_con = np.concatenate([values, values_new])
np.shape(merged_data_con)
````

````
merged_data_vs = np.vstack((values, values_new))
np.shape(merged_data_vs)
````

**np.concatenate()とnp.vstack()の違い**<br>
* numpy.concatenateはすでに存在するaxis上で配列を結合します。画像を横や縦に並べたりするイメージ
* numpy.vstackはnumpy配列を垂直方向に結合します axis=0の結合

**参考**<br>
hstack()　水平方向に結合する　要はaxis=1の結合<br>
dstack()　深さ方向に結合する　要はaxis=2の結合<br>
